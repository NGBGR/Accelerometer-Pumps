```python
#importing all necessary libraries

from nptdms import TdmsFile
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from statsmodels.tsa.stattools import adfuller

%matplotlib inline
```

```python
#reading in ex1.tdms file

tdms_file = TdmsFile("ex1.tdms")
tdms_groups = tdms_file.groups()
tdms_Untitled = tdms_file.group_channels("Untitled")
tdms_Untitled
```

```python
#Generating data arrays of features

for group in tdms_file.groups():
    Accel_LW236869_Pump_X = tdms_file.object(group, 'Accel LW236869 Pump X').data
    Accel_LW236869_Pump_Y = tdms_file.object(group, 'Accel LW236869 Pump Y').data
    Accel_LW236869_Pump_Z = tdms_file.object(group, 'Accel LW236869 Pump Z').data
    Accel_LW236867_Motor_X = tdms_file.object(group, 'Accel LW236867 Motor X').data
    Accel_LW236867_Motor_Y = tdms_file.object(group, 'Accel LW236867 Motor Y').data
    Accel_LW236867_Motor_Z = tdms_file.object(group, 'Accel LW236867 Motor Z').data
    Temp0 = tdms_file.object(group, 'Temp0').data
    Temp1 = tdms_file.object(group, 'Temp1').data
    Suction_Pressure = tdms_file.object(group, 'Suction Pressure').data
    Discharge_Pressure = tdms_file.object(group, 'Discharge Pressure').data
    Flow_Meter = tdms_file.object(group, 'Flow Meter').data
```

```python
#Printing Data Type and Shape of Features

print('Data Type and Shape of Temp0:', type(Temp0), Temp0.shape)
print('Data Type and Shape of Temp1:', type(Temp1), Temp1.shape)
print('Data Type and Shape of Flow_Meter:', type(Flow_Meter) , Flow_Meter.shape)
print('Data Type and Shape of Discharge_Pressure:', type(Discharge_Pressure), Discharge_Pressure.shape)
print('Data Type and Shape of Suction_Pressure:', type(Suction_Pressure), Suction_Pressure.shape)
print('Data Type and Shape of Accel_LW236867_Motor_X:', type(Accel_LW236867_Motor_X), Accel_LW236867_Motor_X.shape)
print('Data Type and Shape of Accel_LW236867_Motor_Y:', type(Accel_LW236867_Motor_Y), Accel_LW236867_Motor_Y.shape)
print('Data Type and Shape of Accel_LW236867_Motor_Z:', type(Accel_LW236867_Motor_Z), Accel_LW236867_Motor_Z.shape)
print('Data Type and Shape of Accel_LW236869_Pump_X:', type(Accel_LW236869_Pump_X), Accel_LW236869_Pump_X.shape)
print('Data Type and Shape of Accel_LW236869_Pump_Y:', type(Accel_LW236869_Pump_Y), Accel_LW236869_Pump_Y.shape)
print('Data Type and Shape of Accel_LW236869_Pump_Z:', type(Accel_LW236869_Pump_Z), Accel_LW236869_Pump_Z.shape)
```

Augmented Dicky Fuller Unit Root Test of Stationarity for Accel_LW236867_Motor_X

```python
series = df3['Accel_LW236867_Motor_X']
X = series.values
result = adfuller(X)
```

```python
#Printing out the ADF Test Statistic, p-value and Critical Values of the Test

print('The ADF Statistic is: %f' % result[0])
print('The p-value is: %f' % result[1])
print('The Critical Values are:')
for key, value in result[4].items():
    print('\t%s: %.3f' % (key, value))
```
